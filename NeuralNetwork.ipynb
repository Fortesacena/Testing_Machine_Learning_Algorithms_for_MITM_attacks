```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = pd.read_csv("Dataset/WIFIDATA.csv",nrows=200)
#df = pd.read_csv('Dataset ( INserto file ).csv')
```

```python
df
```

```python
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = df.fillna("Unknown")
le = LabelEncoder()
df["ssid_name"] = le.fit_transform(df["ssid_name"])
df["phy_type_id"] = le.fit_transform(df["phy_type_id"])
df["capabilities"] = le.fit_transform(df["capabilities"])
df["channel_center_freq_khz"] = le.fit_transform(df["channel_center_freq_khz"])
df["connection_mode"] = le.fit_transform(df["connection_mode"])
df["authentication"] = le.fit_transform(df["authentication"])
df["encryption"] = le.fit_transform(df["encryption"])
df["vendor_name"] = le.fit_transform(df["vendor_name"])
df["bssid"] = le.fit_transform(df["bssid"])
df["mac"] = le.fit_transform(df["mac"])
```

```python
X = df[["ssid_name", "phy_type_id", "capabilities", "channel_center_freq_khz", "connection_mode",
        "authentication", "encryption", "vendor_name", "bssid", "mac"]]
y = df["result"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from tensorflow.keras.utils import to_categorical
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
y_pred = model.predict(X_test)
y_pred = y_pred.argmax(axis=-1)
```

```python
try:
    prec = precision_score(y_test, y_pred, zero_division=1)
except ZeroDivisionError:
    prec = 0.0
try:
    rec = recall_score(y_test, y_pred)
except ZeroDivisionError:
    rec = 0.0
try:
    f1 = f1_score(y_test, y_pred)
except ZeroDivisionError:
    f1 = 0.0
acc = accuracy_score(y_test, y_pred)

```

```python
print("Accuracy: {:.2f}%".format(acc*100))
print("Precision: {:.2f}%".format(prec*100))
```
