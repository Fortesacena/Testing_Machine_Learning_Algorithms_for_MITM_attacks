```python
import pandas as pd
import matplotlib.pyplot as plt


decision_tree_results = {
    "Algorithm": "Decision Tree",
    "Accuracy": 0.95,
}


logistic_regression_results = {
    "Algorithm": "Logistic Regression",
    "Accuracy": 0.95,
}



svm_results = {
    "Algorithm": "SVM",
    "Accuracy": 0.925,
  
}

neutralnetwork_results = {
    "Algorithm": "Neutral Network",
    "Accuracy": 0.325,
  
}


results_df = pd.DataFrame([decision_tree_results, logistic_regression_results, svm_results, neutralnetwork_results])

# Visualize comparison
plt.figure(figsize=(10, 6))
plt.barh(results_df["Algorithm"], results_df["Accuracy"], color='skyblue')
plt.xlabel('Accuracy')
plt.title('Algorithm Comparison - Accuracy')
plt.show()
```
